```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://ucdavisaggies.com/sports/mens-basketball/stats"

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
table = soup.find_all('table')[5] #game by game team stats
print(table) 
```

```python
#Extract the table headers

headers = table.find_all('th')
print(headers)
num_headers = len(headers)
print("Number of headers: ", num_headers)
    
#clean up headers 
table_headers = [title.text.strip() for title in headers]
print(table_headers)
```

```python
#Extract data for the columns
column_data = table.find_all('tr')
df = pd.DataFrame(columns = table_headers)

for row in column_data[2:]: 
    row_data = row.find_all('td')
    print(row_data)
    
        # Extract and clean the text properly
    row_data = [data.get_text(separator=" ").strip() for data in row_data]

    # Ensure that each cell is a string (not a list)
    row_data = [item if isinstance(item, str) else " ".join(item) for item in row_data]

        
    length = len(df)
    df.loc[length] = row_data
    
```

```python
#Export to a csv file
df.to_csv(r'/Users/ryanuyeki/Downloads/asa/team_game_stats.csv', index=False)
```
